```python
from textblob import TextBlob
```

```python
# Obtaining number of days stayed
days_stayed = []
for count, tag in enumerate(hoteldf["Tags"]):
    if (count % 10000) == 0:
        print(count, end=" ")
    review = TextBlob(tag)
    for i, x in enumerate(review.words):
        if "Stayed" in x:
            days_stayed.append(int(review.words[i+1]))
            break
print("Done")
```

```python
# Obtaining type of Stay (Leisure/Business/Unknown)
stay_type = []
counter1 = np.zeros(len(hoteldf["Tags"]))
for count, tag in enumerate(hoteldf["Tags"]):
    if (count % 10000) == 0:
        print(count, end=" ")
    review = TextBlob(tag)
    for i, x in enumerate(review.words):
        if ((x == "Leisure") or (x == "Business")):
            stay_type.append(x)
            counter1[count] = 1
            break
        if ((counter1[count] == 0) and (review.words[-1] != "Leisure") and (review.words[-1] != "Business")):
            stay_type.append("Unknown")
            counter1[count] = 1
            break
print("Done")
```

```python
print(len(stay_type))
print(pd.Series(stay_type).value_counts()/len(stay_type))
```

```python
# Understanding distribution of purpose of stays
ax = plt.gca()
pd.Series(stay_type).value_counts().plot.bar()
plt.xlabel("Purpose of Stay")
plt.ylabel("Count")
plt.title("Distribution of Purpose of Stays")
ax.set_xticklabels(ax.get_xticklabels(), rotation = 0, ha="center", fontsize=10)
plt.show()
```
